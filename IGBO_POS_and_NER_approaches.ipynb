```python

# Converting data into features
def word2features(sent, i):
    word = sent[i][0]
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
    }

    if i > 0:
        prev_word = sent[i - 1][0]
        features.update({
            '-1:word.lower()': prev_word.lower(),
            '-1:word.istitle()': prev_word.istitle(),
            '-1:word.isupper()': prev_word.isupper(),
        })
    else:
        features['BOS'] = True

    if i < len(sent) - 1:
        next_word = sent[i + 1][0]
        features.update({
            '+1:word.lower()': next_word.lower(),
            '+1:word.istitle()': next_word.istitle(),
            '+1:word.isupper()': next_word.isupper(),
        })
    else:
        features['EOS'] = True

    return features

#convert sentences to features
def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

#convert sentences to labels
def sent2labels(sent):
    return [label for _, label in sent]

# Preparing data
X = [sent2features(sent) for sent in sample_data_igbo_test]
y = [sent2labels(sent) for sent in sample_data_igbo_test]

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

#predict with default crf model
y_pred = crf.predict(X_test)

# Flatten y_test and y_pred
y_test_flat = [label for sent_labels in y_test for label in sent_labels]
y_pred_flat = [label for sent_labels in y_pred for label in sent_labels]


# hyperparameter grid to find best parameters
param_grid = {
    'c1': [0.005,0.01,0.02, 0.05, 0.1,0.5, 1],
    'c2': [0.005,0.01,0.02, 0.05, 0.1,0.5, 1],
    'algorithm': ['lbfgs'],
    'max_iterations': [ 200, 300,400,500],
}

best_score = 0
best_params = None


# Go through hyperparameter grid combinations
for c1 in param_grid['c1']:
    for c2 in param_grid['c2']:
        for max_iter in param_grid['max_iterations']:
            # Create a CRF model with current hyperparameters
            crf = CRF(c1=c1, c2=c2, algorithm='lbfgs', max_iterations=max_iter)

            # Train the CRF model
            crf.fit(X_train, y_train)

            # Make predictions on the test set
            y_pred = crf.predict(X_test)

            # Evaluate performance
            score = f1_score(y_test_flat, y_pred_flat, average='macro')

            # Check if current hyperparameters result in a better score
            if score > best_score:
                best_score = score
                best_params = {'c1': c1, 'c2': c2, 'algorithm': 'lbfgs', 'max_iterations': max_iter}

# Print the best hyperparameters
print("Best Hyprparameters:", best_params)
```

```python
# Create the best CRF model with the best parameters
best_crf = sklearn_crfsuite.CRF(
    algorithm=best_params['algorithm'],
    c1=best_params['c1'],
    c2=best_params['c2'],
    max_iterations=best_params['max_iterations']
)

# Train the model on the entire training set
best_crf.fit(X_train, y_train)

# Make predictions on the test set
y_pred_best = best_crf.predict(X_test)
# Flatten y_test and y_pred
y_test_flat = [label for sent_labels in y_test for label in sent_labels]
y_pred_flat = [label for sent_labels in y_pred_best for label in sent_labels]

# Print classification report on accuracoes
print("Classification Report:")
print(classification_report(y_test_flat, y_pred_flat))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from simpletransformers.ner import NERModel,NERArgs

# Read sample data (Bible Book of Jon chapter 1 with reedited copy of itself for better results)
dataset = pd.read_excel('ner dataset.xlsx', sheet_name='Sheet1')  # pd.read_csv("ner_dataset.csv", encoding="latin1" )
dataset = dataset.fillna(method="ffill")
dataset.head(10)
```

```python
# Encode sentence labels
dataset["Sentence #"] = LabelEncoder().fit_transform(dataset["Sentence #"])
dataset.head(10)

# Rename columns for better readability
dataset.rename(columns={"Sentence #": "sentence_id", "Word": "words", "Tag": "labels"}, inplace=True)
dataset["labels"] = dataset["labels"].str.upper()

# Split data into training and testing sets
X = dataset[["sentence_id", "words"]]
Y = dataset["labels"]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Build train data and test data
train_data = pd.DataFrame({"sentence_id": x_train["sentence_id"], "words": x_train["words"], "labels": y_train})
test_data = pd.DataFrame({"sentence_id": x_test["sentence_id"], "words": x_test["words"], "labels": y_test})

# Importing from simpletransformers without revealing the model
from simpletransformers.ner import NERModel, NERArgs

# Unique labels
labels = dataset["labels"].unique().tolist()

# Model configuration
args = NERArgs()
args.num_train_epochs = 41
args.learning_rate = 1e-5
args.overwrite_output_dir = True
args.train_batch_size = 32
args.eval_batch_size = 32

# Initialize NERModel
model = NERModel('bert', 'bert-base-cased', labels=labels, args=args)

# Make a prediction before any training with the labels of our interest
prediction, model_output = model.predict(["Jisus zara, si ya, Gì onwe-gi bu onye-ozizí Israel, ma i maghi ihe ndia?"])
prediction
```

```python
#train model
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
#evaluate model
result, model_outputs, preds_list = model.eval_model(test_data)
result
```
