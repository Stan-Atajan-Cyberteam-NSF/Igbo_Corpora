# Akuko pages

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
data2=[]
# Define the URL of the starting page
start_url =  "https://www.bbc.com/igbo"
# Keep track of the URLs we've already visited
visited_urls = set()

# Define a function to scrape URLs from a page
def scrape_urls(url):
    # Send an HTTP request to the server
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML response
        soup = BeautifulSoup(response.text, "html.parser")
        body_text = soup.get_text()
       
################
        # Find all the paragraphs in the HTML content
        paragraphs = soup.find_all("p")

        # Extract the text from the paragraphs
        text = [p.text for p in paragraphs]

        # Join the text into a single string
        text = " ".join(text)




################
        # Add the body text and URL to the dataframe
        data2.append({"url": url, "body_text": body_text, 'paragraph': text})

        # Find all links on the page
        links = soup.find_all("a")
        soup.decode('utf-8', 'ignore')

        # Loop through each link
        for link in links:
            # Extract the link's URL
            link_url = link.get("href")

            # Check if the link URL is not None
            if link_url:
                # Check if the link is internal to the domain
                if link_url.startswith(start_url):
                    # Check if the link has not been visited yet
                    if link_url not in visited_urls:
                        # Add the link to the list of visited URLs
                        visited_urls.add(link_url)

                        # Recursively scrape the linked page
                        scrape_urls(link_url)

# Start the scraping process
scrape_urls(start_url)
df2 = pd.DataFrame(data2)

# Print the list of visited URLs
print(len(visited_urls))
```

```python
#@title saving Akuko corpus as Akuko.csv under the folder on the left panel
df2[['url','paragraph']].to_csv('Akuko.csv',encoding='utf-8-sig')
```

```python
# convert corpus to single lines
import pandas as pd

data = pd.read_csv('Egwuregwu.csv')

data.head()


```

```python
# loop each paragraph and create as many sentences it has to single lines
# go through each row 
# split the text by the ending character,
# assigne the split to new row based on the loop of number of splits
#sentences_df = pd.DataFrame(columns=['url', 'Sentence'])


def converter(file):
  sentences_df = pd.DataFrame(columns=['url', 'Sentence'])

  # loop through the original dataframe and split sentences
  for idx, row in file.iterrows():
      index = row['url']
      text = row['paragraph']
      sentences = text.split('. ') # split based on '.' character
      for sentence in sentences:
          # append each sentence to the new dataframe with the same index
          sentences_df = sentences_df.append({'url': index, 'Sentence': sentence.strip()}, ignore_index=True)
          return(sentences_df)


converter(data.head(3))
```

```python
def split_sentences_with_char(df, col_name, char):
    sentences = []
    indices = []
    urls = []
    for index, row in df.iterrows():
        sentence_list = row[col_name].split(char)
        for sentence in sentence_list:
            sentences.append(sentence + char)
            indices.append(index)
            urls.append(row['url'])
    new_df = pd.DataFrame({'Index': indices, 'Sentence': sentences, 'URL': urls})
    return new_df



egwuregwu_corpus = split_sentences_with_char(data, 'paragraph', '. ')
egwuregwu_corpus 
```
